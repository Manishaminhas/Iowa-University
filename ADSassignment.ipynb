```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing data from file
Iowa= pd.read_csv('Iowa.csv')
```

```python
#checking how data looks
Iowa.head()
```

```python
#plotting bar chart
sns.barplot(Iowa['Semester'], Iowa['Total'])
```

```python
#plotting pie chart
Iowa.groupby(['Semester']).sum().plot(kind='pie', subplots=True, shadow = True,startangle=0.9,
figsize=(20, 20), autopct='%1.1f%%')
```

```python
# form dataframe from data
df = pd.DataFrame(Iowa, columns=["Semester", "Ungrad", "Prof", "Grad", "Post-Doc"])
 
# plot multiple columns from dataframe
df.plot(x="Semester", y=["Ungrad", "Prof", "Grad", "Post-Doc"],
        kind="line", figsize=(10, 10))
 
# display plot
mp.show()
```
